#  Diabetes Diagnosis Model Documentation

## Project Overview:
This project aims to build a predictive model for diagnosing diabetes using the **Behavioral Risk Factor Surveillance System (BRFSS)** dataset, which contains survey responses about health-related behaviors and chronic conditions. The target variable is `Diabetes_binary`, where 0 represents non-diabetic individuals, and 1 represents prediabetic or diabetic individuals.

### Project Focus:
- **Accurately predict** whether an individual has diabetes.
- **Minimize false negatives**, ensuring as many diabetes cases as possible are detected (prioritizing recall).
- Compare multiple machine learning models and apply techniques like **SMOTE** and **threshold tuning** to optimize performance.

## Data Preprocessing:

### Loading and Inspection:
- The dataset was loaded and inspected for missing values, with none found.

### Feature Scaling:
- Features were standardized to ensure optimal performance of the machine learning models.

### Addressing Class Imbalance:
- The dataset was imbalanced, with fewer diabetes cases. To address this, **SMOTE** (Synthetic Minority Over-sampling Technique) was applied to generate synthetic samples of the minority class (diabetes). This helped improve model performance by providing balanced data for training.

## Can Survey Questions from the CDC Accurately Predict Diabetes?
Yes, the CDC’s BRFSS survey questions provide accurate predictions of diabetes risk. After applying machine learning models and addressing class imbalance using SMOTE, we achieved a recall of 95% with the **Random Forest** model (with threshold tuning), demonstrating the effectiveness of the survey data.

### Recommendation for Additional Features:
- Incorporating **medical data** such as blood test results (e.g., glucose levels, HbA1c) and **family history** could improve accuracy further.
- Additional lifestyle factors, such as **sleep patterns** and **stress levels**, could provide a more holistic view of diabetes risk.

## Model Development:
We evaluated three machine learning models for predicting diabetes:
- **Logistic Regression**
- **Random Forest Classifier**
- **K-Nearest Neighbors (KNN)**

### Model Results (without SMOTE):
- Initial models achieved high accuracy but struggled with recall for class 1 (diabetes), missing a significant portion of actual diabetes cases.

### Handling Imbalance with SMOTE:
- Applying **SMOTE** significantly improved recall for all models, especially Random Forest:
  - **Logistic Regression**: Recall improved from 18% to 78%.
  - **Random Forest**: Recall improved from 18% to 89%.
  - **K-Nearest Neighbors (KNN)**: Recall improved from 21% to 95%.

### Key Risk Factors for Diabetes:
Based on feature importance from the **Random Forest** model, the most predictive factors for diabetes were:
1. **High Blood Pressure (HighBP)**: Strong correlation with diabetes risk.
2. **BMI (Body Mass Index)**: Higher BMI is associated with increased risk.
3. **Age**: Older individuals are at higher risk.
4. **General Health (GenHlth)**: Poorer general health is linked to diabetes.
5. **Physical Activity (PhysActivity)**: Lack of physical activity correlates with higher risk.

### Can a Subset of Risk Factors Accurately Predict Diabetes?
Yes, using a subset of the top five features (HighBP, BMI, Age, GenHlth, PhysActivity), models could still achieve good performance, particularly with **Random Forest**. However, the full set of features provides a more robust model capable of capturing complex relationships.

## Model Comparison and Evaluation:

### Random Forest (Final Model):
- **Accuracy**: 91.76%
- **Recall for class 1 (diabetes)**: 95% (after threshold tuning and SMOTE)
- **Precision for class 1**: 84%

**Reason for Performance**: Random Forest's ability to handle complex interactions between features allowed it to outperform other models, especially in minimizing false negatives, which is crucial for medical diagnosis.

### Logistic Regression:
- **Accuracy**: 75.47%
- **Recall for class 1**: 78% (after SMOTE)
- **Precision for class 1**: 74%

**Reason**: While Logistic Regression improved after SMOTE, it struggled with non-linear relationships compared to Random Forest.

### K-Nearest Neighbors (KNN):
- **Accuracy**: 83.99%
- **Recall for class 1**: 95%
- **Precision for class 1**: 78%

**Reason**: KNN achieved high recall but lower precision, resulting in more false positives. Additionally, its performance was impacted by the large dataset size.

## Threshold Tuning:

### Rationale:
In medical diagnosis, minimizing false negatives is crucial. Therefore, threshold tuning was used to optimize the balance between precision and recall.

### Threshold Tuning Results for Random Forest:

| Threshold | Precision | Recall | F1-Score | Accuracy |
| --------- | --------- | ------ | -------- | -------- |
| 0.3       | 84%       | 95%    | 89%      | 88%      |
| 0.4       | 89%       | 92%    | 91%      | 91%      |
| 0.5       | 93%       | 90%    | 92%      | 92%      |
| 0.6       | 96%       | 87%    | 91%      | 92%      |
| 0.7       | 98%       | 84%    | 91%      | 91%      |

### Final Decision:
- **Threshold = 0.3** was selected to maximize recall (95%), ensuring the model captures most diabetes cases, with an acceptable trade-off in precision.

## Conclusion:
The **Random Forest** model, with a threshold of **0.3**, was chosen as the final model for deployment. This threshold ensures 95% recall, crucial for minimizing missed diagnoses in diabetes prediction. With SMOTE applied to balance the dataset, the model achieved excellent performance, with an overall accuracy of **91.76%**.

## Step 3: Feature Engineering

### Why Feature Engineering?
- **Enhance Predictive Power:**  
  By creating new features or transforming existing ones, we can improve the ability of our machine learning models to make accurate predictions.
  
- **Capture Relationships:**  
  Sometimes, interactions between features (such as mental and physical health) can reveal important patterns that aren’t obvious from the original features alone.

---

### What We Can Do:

#### 1. Create Interaction Features
We’ll create a health index that combines mental health (`MentHlth`), physical health (`PhysHlth`), and general health (`GenHlth`) to capture the overall health condition of individuals.



```python
# Create interaction feature (Health Index)
df['HealthIndex'] = df['GenHlth'] * df['MentHlth'] * df['PhysHlth']

# Create polynomial features (BMI^2, Age^2, MentHlth^2)
df['BMI_Squared'] = df['BMI'] ** 2
df['Age_Squared'] = df['Age'] ** 2
df['MentHlth_Squared'] = df['MentHlth'] ** 2

# Drop features with low correlation (Fruits, Veggies, HvyAlcoholConsump)
df = df.drop(columns=['Fruits', 'Veggies', 'HvyAlcoholConsump'])

# Check the updated dataset
print(df.head())
print("New shape of dataset:", df.shape)
```

# Step 3: Feature Engineering Recap

We've successfully:

- **Created interaction features** like `HealthIndex` (combining general health, mental health, and physical health).
- **Generated polynomial features** for `BMI`, `Age`, and `MentHlth` to capture potential non-linear relationships.
- **Removed low-importance features** based on correlation (`Fruits`, `Veggies`, `HvyAlcoholConsump`).
- The dataset now has **23 features** and is ready for the modeling phase.

# Step 4: Model Selection

Next, we will:

- **Split the dataset** into training and testing sets.
- **Choose three machine learning models** for comparison. Since we're avoiding deep learning, we'll use:
  - **Logistic Regression**: A simple linear model that works well for binary classification tasks.
  - **Random Forest**: A tree-based model that can handle non-linear relationships and interactions between features.
  - **Support Vector Machine (SVM)**: A robust classifier that can separate classes with a hyperplane, often used for classification tasks with complex decision boundaries.


```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_auc_score, roc_curve
import matplotlib.pyplot as plt

# Split the data into features (X) and target (y)
X = df.drop('Diabetes_binary', axis=1)
y = df['Diabetes_binary']

# Standardize the feature data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Initialize the models
logreg = LogisticRegression(max_iter=2000)  # Increased max_iter
rf = RandomForestClassifier(random_state=42)
knn = KNeighborsClassifier(n_neighbors=5)

# Train and evaluate Logistic Regression
logreg.fit(X_train, y_train)
y_pred_logreg = logreg.predict(X_test)
print("Logistic Regression Accuracy:", accuracy_score(y_test, y_pred_logreg))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_logreg))
print("Classification Report:\n", classification_report(y_test, y_pred_logreg))

# Train and evaluate Random Forest
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)
print("Random Forest Accuracy:", accuracy_score(y_test, y_pred_rf))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_rf))
print("Classification Report:\n", classification_report(y_test, y_pred_rf))

# Train and evaluate KNN
knn.fit(X_train, y_train)
y_pred_knn = knn.predict(X_test)
print("K-Nearest Neighbors (KNN) Accuracy:", accuracy_score(y_test, y_pred_knn))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_knn))
print("Classification Report:\n", classification_report(y_test, y_pred_knn))

# ROC-AUC for Logistic Regression, Random Forest, and KNN
y_prob_logreg = logreg.predict_proba(X_test)[:, 1]
y_prob_rf = rf.predict_proba(X_test)[:, 1]
y_prob_knn = knn.predict_proba(X_test)[:, 1]

roc_auc_logreg = roc_auc_score(y_test, y_prob_logreg)
roc_auc_rf = roc_auc_score(y_test, y_prob_rf)
roc_auc_knn = roc_auc_score(y_test, y_prob_knn)

# ROC curves
fpr_logreg, tpr_logreg, _ = roc_curve(y_test, y_prob_logreg)
fpr_rf, tpr_rf, _ = roc_curve(y_test, y_prob_rf)
fpr_knn, tpr_knn, _ = roc_curve(y_test, y_prob_knn)

# Plot the ROC curves for all models
plt.figure(figsize=(10,6))
plt.plot(fpr_logreg, tpr_logreg, label=f"Logistic Regression (AUC = {roc_auc_logreg:.2f})")
plt.plot(fpr_rf, tpr_rf, label=f"Random Forest (AUC = {roc_auc_rf:.2f})")
plt.plot(fpr_knn, tpr_knn, label=f"KNN (AUC = {roc_auc_knn:.2f})")
plt.plot([0, 1], [0, 1], 'k--', label="Random Chance (AUC = 0.50)")
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curves for Logistic Regression, Random Forest, and KNN')
plt.legend(loc="lower right")
plt.show()

# Print ROC-AUC scores
print(f"Logistic Regression ROC-AUC Score: {roc_auc_logreg:.2f}")
print(f"Random Forest ROC-AUC Score: {roc_auc_rf:.2f}")
print(f"KNN ROC-AUC Score: {roc_auc_knn:.2f}")
```

### Should We Use SMOTE?

Given the low recall for diabetes across all models, **SMOTE** (Synthetic Minority Over-sampling Technique) could help improve the recall for class 1 (diabetes). This imbalanced dataset requires SMOTE to help the models learn better patterns for the minority class, potentially improving recall without sacrificing too much precision.

#### Why SMOTE is Necessary:
- The dataset is imbalanced, as reflected in the confusion matrices and classification reports, where class 0 (non-diabetes) dominates.
- All models have low recall for diabetes, indicating that they are not identifying a significant portion of diabetes cases.
- **SMOTE** will create synthetic examples of the minority class, encouraging the models to pay more attention to class 1 (diabetes).

---

### Next Steps:
1. **Apply SMOTE** to balance the dataset.
2. **Retrain the models** and evaluate if recall for diabetes improves.
3. After applying SMOTE, **check the trade-off between recall and precision** to ensure overall model performance is still acceptable.


```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_auc_score, roc_curve
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# Split the data into features (X) and target (y)
X = df.drop('Diabetes_binary', axis=1)
y = df['Diabetes_binary']

# Standardize the feature data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply SMOTE to balance the dataset
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_scaled, y)

# Split the resampled dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)

# Initialize the models
logreg = LogisticRegression(max_iter=2000)
rf = RandomForestClassifier(random_state=42)
knn = KNeighborsClassifier(n_neighbors=5)

# Train and evaluate Logistic Regression
logreg.fit(X_train, y_train)
y_pred_logreg = logreg.predict(X_test)
print("Logistic Regression Accuracy (with SMOTE):", accuracy_score(y_test, y_pred_logreg))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_logreg))
print("Classification Report:\n", classification_report(y_test, y_pred_logreg))

# Train and evaluate Random Forest
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)
print("Random Forest Accuracy (with SMOTE):", accuracy_score(y_test, y_pred_rf))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_rf))
print("Classification Report:\n", classification_report(y_test, y_pred_rf))

# Train and evaluate KNN
knn.fit(X_train, y_train)
y_pred_knn = knn.predict(X_test)
print("K-Nearest Neighbors (KNN) Accuracy (with SMOTE):", accuracy_score(y_test, y_pred_knn))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_knn))
print("Classification Report:\n", classification_report(y_test, y_pred_knn))

# ROC-AUC for Logistic Regression, Random Forest, and KNN
y_prob_logreg = logreg.predict_proba(X_test)[:, 1]
y_prob_rf = rf.predict_proba(X_test)[:, 1]
y_prob_knn = knn.predict_proba(X_test)[:, 1]

roc_auc_logreg = roc_auc_score(y_test, y_prob_logreg)
roc_auc_rf = roc_auc_score(y_test, y_prob_rf)
roc_auc_knn = roc_auc_score(y_test, y_prob_knn)

# ROC curves
fpr_logreg, tpr_logreg, _ = roc_curve(y_test, y_prob_logreg)
fpr_rf, tpr_rf, _ = roc_curve(y_test, y_prob_rf)
fpr_knn, tpr_knn, _ = roc_curve(y_test, y_prob_knn)

# Plot the ROC curves for all models
plt.figure(figsize=(10,6))
plt.plot(fpr_logreg, tpr_logreg, label=f"Logistic Regression (AUC = {roc_auc_logreg:.2f})")
plt.plot(fpr_rf, tpr_rf, label=f"Random Forest (AUC = {roc_auc_rf:.2f})")
plt.plot(fpr_knn, tpr_knn, label=f"KNN (AUC = {roc_auc_knn:.2f})")
plt.plot([0, 1], [0, 1], 'k--', label="Random Chance (AUC = 0.50)")
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curves for Logistic Regression, Random Forest, and KNN with SMOTE')
plt.legend(loc="lower right")
plt.show()

# Print ROC-AUC scores
print(f"Logistic Regression ROC-AUC Score (with SMOTE): {roc_auc_logreg:.2f}")
print(f"Random Forest ROC-AUC Score (with SMOTE): {roc_auc_rf:.2f}")
print(f"KNN ROC-AUC Score (with SMOTE): {roc_auc_knn:.2f}")
```
